## Sierpinski Triangle

#### First Transformation
$$x_{n+1}=0.5x_n$$  
$$y_{n+1} = 0.5y_n$$ 

#### Second Transformation
$$x_{n+1} = 0.5x_n+ 0.5$$  
$$y_{n+1} = 0.5y_n+ 0.5$$  

#### Third Transformation
$$x_{n+1} = 0.5x_{n+1}$$  
$$y_{n+1} = 0.5y_n$$ 

## FIrst we should import the relevant libraries

```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1,y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1,y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y
    return x1,y1

transformations = [trans_1,trans_2,trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0

for i in range(1000000):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
```

```python
plt.rc('figure', figsize=(16,16))
```

```python
plt.plot(a1,b1,"o")
plt.savefig("SierpinskiTriangle.png")
```
